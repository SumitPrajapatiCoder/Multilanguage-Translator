# 🚀 Project Setup & Run Instructions

## 🔹 1️⃣ AI Service (FastAPI)

Create a .env file inside the ai-service folder:

```python
CLOUDINARY_NAME=<your_cloudinary_name>
CLOUDINARY_API_KEY=<your_cloudinary_api_key>
CLOUDINARY_API_SECRET=<your_cloudinary_api_secret>
```

```python
cd ai-service
```

### Create Virtual Environment (First time only)

```python
python -m venv venv
```

### Activate Virtual Environment (Windows)

```python
venv\Scripts\activate
```

### Install Requirements

```python
pip install -r requirements.txt
```

### Run FastAPI Server

```python
python -m uvicorn app:app --port 8000 --reload
```

## 🔹 2️⃣ Backend (Node.js / Express)

Create a .env file inside the backend folder and configure:

```python
PORT=5000
NODE_ENV=development

MONGO_URI=<your_mongodb_connection_string>

JWT_SECRET=<your_jwt_secret>

CLOUDINARY_NAME=<your_cloudinary_name>
CLOUDINARY_API_KEY=<your_cloudinary_api_key>
CLOUDINARY_API_SECRET=<your_cloudinary_api_secret>

OPENAI_API_KEY=<your_openai_api_key>

GOOGLE_APPLICATION_CREDENTIALS=<path_to_google_credentials_json>

LIVEKIT_URL=<your_livekit_url>
LIVEKIT_API=<your_livekit_api_key>
LIVEKIT_API_SECRET=<your_livekit_api_secret>

FRONTEND_URL=http://localhost:5173

LIVEKIT_ROOM_NAME=<default_room_name>
LIVEKIT_MAX_PARTICIPANTS=10
```

```python
cd backend
```

### Install Dependencies

```python
npm install
```

### Run Backend Server

```python
npm run dev
```

## 🔹 3️⃣ Frontend (React / Vite)

Create a .env file inside the frontend folder and add:

```python
VITE_LIVEKIT_API_KEY=<your_livekit_api_key>
```

```python
cd frontend
```

### Install Dependencies

```python
npm install
```

### Run Frontend

```python
npm run sumit
```
